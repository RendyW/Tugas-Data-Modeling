```python
# menampilkan informasi untuk mengetahui kolom/atribut data yang memiliki data null (empty) dalam barisnya

print(dataset.isnull().sum())
```

```python
# menghapus kolom yang memiliki data null

new_dataset = dataset.dropna(axis = 1)
new_dataset.info()
```

```python
# menghapus baris yang memiliki data null

new_dataset2 = dataset.dropna(axis=0)
new_dataset2.info()
```

```python
# mengisi data null dengan nilai 0

default_dataset = dataset.fillna(0)
default_dataset.info()
```

```python
# Cara lain mengisi data ke hanya kolom yang memiliki nilai NULL

default_dataset_spec = dataset.copy()
default_dataset_spec['Description'] = default_dataset_spec['Description'].fillna('no description')
```

```python
# Mengisi bagian null pada kolom 'Customer ID' dengan 0

default_dataset_spec['Customer ID'] = default_dataset_spec['Customer ID'].fillna(0)
```

```python
# menampilkan informasi data

default_dataset_spec.info()
```

```python
# Membuat kolom baru bernama 'mean_dataset'
import numpy as np

mean_dataset = dataset.copy()
mean_dataset['mean_quantity'] = np.mean
mean_dataset.info()
```

```python
# Mengisi kolom baru bernama 'mean_dataset'

mean_dataset['mean_quantity'] = mean_dataset['mean_quantity'].fillna(mean_dataset['Quantity'].mean())
mean_dataset.info()
mean_dataset
```

```python
# Mengisi data null dengan berdasarkan kondisi tertentu

data_cond = dataset.copy()

data_cond['Description'] = np.where(np.logical_and(data_cond['Description'].isnull(), data_cond['Quantity'] > 10), 'bulk',
                                    np.where(data_cond['Description'].isnull(), 'not clear', data_cond['Description']))
data_cond.info()
```

```python
# memastikan data sudah terisi sesuai kondisi yang ditentukan

data_cond[data_cond['Description'] == 'bulk']
```

```python
data_cond[data_cond['Description'] == 'not clear']
```

### Regular Expression (Regex)

```python
#import library Regex

import re
```

Function pada library re:
- re.findall() mengembalikan list kata yang ditemukan dalam sebuah teks atau kalimat.
- re.search() untuk menjadi kata atau huruf tertentu, hanya mengembalikan kata yang pertama kali ditemukan dari semua kata yang mungkin cocok.
- re.split() untuk memisahkan sebuah kata berdasarkan separator yang ditentukan.
- re.sub() untuk mengganti beberapa teks menjadi teks yang lain.

```python
# Menggunakan re.findall()
var = "Contoh teks yang digunakan untuk uji coba function dari library re, pencarian teks, memisah teks, dan mengganti teks"

print(re.findall('teks', var))
```

```python
# Menggunakan re.search()
print(re.search('teks', var))
```

```python
# Menggunakan re.split()
print(re.split(' ', var))
```

### CONTAINS vs. MATCH

notes:
- Untuk mencari data yang memiliki kata mirip/persis gunakan match()
- Untuk mencari data yang memiliki kata tertentu dapat menggunakan contains()

```python
data_contain = dataset['Description'].str.contains('LIGHTS')
print(data_contain)
```

```python
data_match = dataset['Description'].str.match('LIGHTS')
print(data_match)
```

```python
data_match2 = dataset['Description'].str.match('FELICRAFT PRINCESS LOLA DOLL')
print(data_match2)
```

Perbedaan hasil output menggunakan contains() dan match() adalah
- contains() bisa menemukan kata yang sama/percis yang terdapat dalam beberapa rows pada dataset
- match()bisa menemukan data yang memiliki kata tertentu

```python
data_replace = dataset['Description'].str.replace('LIGHTS', 'LAMP')
data_replace
```

```python
# menggunakan split()
data_split = dataset['Description'].str.split()
print(data_split)
```

```python
#menggunakan resplit()

data_resplit = dataset['Description'].str.rsplit()
print(data_resplit)
```
